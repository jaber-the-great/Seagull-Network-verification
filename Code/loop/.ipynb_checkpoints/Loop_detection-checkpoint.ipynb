```python
import itertools

max_id = 1
id_map = dict()

def is_bidirectional_graph(nodes, edges):
    cnt = 0
    for n in nodes:
        if not n in adj.keys():
            #print("node ", n, " has no neighbours")
            cnt += 1
            continue
        for nbr in adj[n]:
            if not nbr in adj.keys() or not n in adj[nbr]:
                #print("node ", n, "neighour ", nbr, " does not have back edge")
                cnt += 1
    return cnt == 0
    
#assign a unique id from node string
def get_node_id(node):
    global max_id

    if not node in id_map.keys():
        n1 = max_id
        id_map[node] = max_id
        #print("assigned new node #", id_map[node], "to ", node)
        max_id += 1
    else:
        n1 = id_map[node]
    return n1       
                    
asfile = open('data/cycle-aslinks.l7.t1.c000065.20080102.txt', 'r')
lines = asfile.readlines()

edges = set()
nodes = set()
adj = dict()
for line in lines:
    if line.startswith('D'):
        vline = line.split()
        if len(vline) > 2:
            n1 = get_node_id(vline[1])
            n2 = get_node_id(vline[2])  
            
            edges.add((n1, n2))
            nodes.add(n1)
            nodes.add(n2)
            
            #create reversed edges for FIB alg
            try:
                adj[n2].insert(n1)
            except:
                adj[n2] = [n1]

print("# of nodes:", len(nodes))
print("# of edges:", len(edges))
print("the graph is", "" if is_bidirectional_graph(nodes, edges) else "not", "an undirected graph" )
print("max node # is", max_id-1)
```

```python
import random

def FIB(adj, target_node):
    visited1 = [] # List to keep track of visited nodes.
    queue = []     #Initialize a queue
    FIB = []
    
    visited1.append(target_node)
    queue.append(target_node)

    while queue:
        s = queue.pop(0) 
        #print (s, end = " ")
        
        if s in adj.keys():
            for neighbour in adj[s]:
              #print("\t",neighbour )
              if neighbour not in visited1:
                tmp = [neighbour,s]
                FIB.append(tmp)
                visited1.append(neighbour)
                queue.append(neighbour)

    return FIB


random_nodes = []
for i in range(6):
    random_nodes.append(random.randint(1, max_id))
    
for end_node in random_nodes:
    fib_graph = FIB(adj, end_node)
    print("node", end_node, "'s FIB", fib_graph)
    print("")
```
