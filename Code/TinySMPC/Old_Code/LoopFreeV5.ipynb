```python
from tinysmpc import VirtualMachine, PrivateScalar, SharedScalar
from datetime import datetime
```

Reading the data in file and create input for FIB generation algorithm(it uses BFS)

```python
f = open("280_AS_LCC_IPv6L.edgelist", "r")
edges = []
for line in f:
    x = line.rstrip()
    edges.append(x.split())

MyGraph = {}
temp = []
for e in edges:
    temp = [int(e[1])]
    if MyGraph.get(int(e[0])):
        adj = MyGraph.get(int(e[0]))
        adj.append(int(e[1]))
        MyGraph[int(e[0])] = adj
    else:
        MyGraph[int(e[0])] = temp
    
    temp = [int(e[0])]
    if MyGraph.get(int(e[1])):
        adj = MyGraph.get(int(e[1]))
        adj.append(int(e[0]))
        MyGraph[int(e[1])] = adj
    else:
        MyGraph[int(e[1])] = temp
```

Using BFS to generate the FIB

```python
import random


policies = ['c','p','i','l'] # cumstomer, provider, peer, local 
FIB1 = []
for key in MyGraph:
    
    temp = MyGraph[key]

    choices = []
    for item in temp:
        r = random.randint(0,3)
        choices.append(policies[r])
        
    theChoice = 0
    prevlevel = 0
    for i in range(len(choices)):
        if choices[i]=='l':
            theChoice = i
            prevlevel =3
        elif choices[i] == 'p':
            if prevlevel <3:
                theChoice = i
                prevlevel =2
    
        elif choices[i] == 'i':
            if prevlevel <2:
                theChoice = i
                prevlevel =1  
        else:
            if prevlevel <1:
                theChoice = i
                prevlevel =0
    edge = [key,temp[theChoice]]

    FIB1.append(edge)
    
```

```python
visited1 = [] # List to keep track of visited nodes.
queue = []     #Initialize a queue
FIB = []
DestID = 1273
def bfs(visited1,graph, node):
  visited1.append(node)
  queue.append(node)

  while queue:
    s = queue.pop(0) 
    print (s, end = " ")

    for neighbour in graph[s]:
      if neighbour not in visited1:
        tmp = [neighbour,s] # for reversed FIB
        FIB.append(tmp)
        visited1.append(neighbour)
        queue.append(neighbour)

# Driver Code
bfs(visited1, MyGraph,DestID)
# print(len(visited))
```

```python
machines = []
nodes = []
IDs = {}
for edge in FIB:
    name = int(edge[0])
    next_hop = int(edge[1])
    vm = VirtualMachine(name)
    machines.append(vm)
    IDs[name] = int(next_hop)
    nodes.append(PrivateScalar(next_hop,vm))
```

```python
shared = []
for i in range(len(nodes)):
    shared.append(nodes[i].share([p1, p2, p3]))
```

```python
start_time = datetime.now()
visited = []
visited.append(DestID)
Flag = True

while Flag:
    Flag = False
    if len(visited) == NumberOfNodes:
        break
    for i in range(0,len(shared)):

        if shared[i].reconstruct(q).value in visited:
            temp = nodes[i].owner.name
#             nodeID = IDs.get(temp)
            if temp not in visited:
                visited.append(temp)
        
# print(visited)
# print(len(visited))
if len(visited) == NumberOfNodes:
    print("It is loop free")
else:
    print("A loop is detected")
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```
