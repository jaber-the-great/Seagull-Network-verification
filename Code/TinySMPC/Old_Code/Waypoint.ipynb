```python
from tinysmpc import VirtualMachine, PrivateScalar, SharedScalar
from datetime import datetime
```

Reading the data in file and create input for FIB generation algorithm(it uses BFS)

```python
f = open("33752AS-LCC_combined.edgelist", "r")
edges = []
for line in f:
    x = line.rstrip()
    edges.append(x.split())

MyGraph = {}
temp = []
for e in edges:
    temp = [int(e[1])]
    if MyGraph.get(int(e[0])):
        adj = MyGraph.get(int(e[0]))
        adj.append(int(e[1]))
        MyGraph[int(e[0])] = adj
    else:
        MyGraph[int(e[0])] = temp
    
    temp = [int(e[0])]
    if MyGraph.get(int(e[1])):
        adj = MyGraph.get(int(e[1]))
        adj.append(int(e[0]))
        MyGraph[int(e[1])] = adj
    else:
        MyGraph[int(e[1])] = temp
```

```python
p1 = VirtualMachine('p1')
p2 = VirtualMachine('p2')
p3 = VirtualMachine('p3')
q = VirtualMachine('Querier')
```

Using BFS to generate the FIB

```python
MyGraph
```

```python
visited1 = [] # List to keep track of visited nodes.
queue = []     #Initialize a queue
FIB = []
DestID = 1273

def bfs(visited1,graph, node):
  visited1.append(node)
  queue.append(node)

  while queue:
    s = queue.pop(0) 
    print (s, end = " ")

    for neighbour in graph[s]:
      if neighbour not in visited1:
        tmp = [neighbour, s]
        FIB.append(tmp)
        visited1.append(neighbour)
        queue.append(neighbour)

# Driver Code
bfs(visited1, MyGraph,DestID)
# print(len(visited))
```

```python
machines = []
nodes = []
IDs = {}
for edge in FIB:
    name = int(edge[1])
    next_hop = int(edge[0])
    vm = VirtualMachine(name)
    machines.append(vm)
    IDs[name] = int(next_hop)
    nodes.append(PrivateScalar(next_hop,vm))
```

```python
shared = []
for i in range(len(nodes)):
    shared.append(nodes[i].share([p1, p2, p3]))
```

Build the path for test

```python
SourceID =34025
pointID = 20965
visited = []
position = SourceID
visited.append(SourceID)
Flag = True
while Flag:
    Flag = False
    for i in range(0,len(shared)):
        if shared[i].reconstruct(q).value == position:
            Flag = True
            temp = nodes[i].owner.name
            nodeID = IDs.get(temp)
            if temp == DestID:
                Flag = False
            position = temp
            visited.append(temp)
print(visited)
```
